```python
import pandas as pd
import pandas_datareader.data as pdr
import matplotlib.pyplot as plt
import datetime
import yfinance as yf
```

```python
data = yf.download("PG", start="1995-01-01")
print(data)
```

Criação de uma coluna para representar o dechamento diario do preço da ação

```python
data["simple_return"] = (data["Adj Close"] / data["Adj Close"].shift(1)) - 1
print(data["simple_return"])
```

```python
data['simple_return'].plot(figsize=(8, 5))
plt.show()
```

RETORNO ANUAL MEDIO POR ANO DESDE 1995 ATÉ ATUALMENTE (2023 - MAIO)

```python
annual_return_mean = data['simple_return'].mean() * 250
print(round(annual_return_mean * 100, 2), "%")
```
